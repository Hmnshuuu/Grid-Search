```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\himan\Music\2.1 UNZIP_ME_FOR_NOTEBOOKS_V4\08-Linear-Regression-Models/Advertising.csv"

# Read the CSV file using the absolute path
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# create X and y
X = df.drop('sales',axis=1)
y = df['sales']

# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)

# scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_net_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,50,100],
              'l1_ratio':[.1,.5,.7,.95,.99,1]}
```

```python
#perform grid search

from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_net_model,
                         param_grid=param_grid,
                         scoring='neg_mean_squared_error',
                         cv=5,verbose=2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```
